```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100, endpoint=True)
print(x)
```

```python
y = x**6 * (1-x)**4
plt.plot(x, y, 'o')
plt.xlabel("theta")
plt.ylabel("L(theta)")
plt.show()
```

```python
index = np.argmax(y)
```

```python
print(x[index])
```
